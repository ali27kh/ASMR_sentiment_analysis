```python
import pandas as pd
import numpy as np
import re
import nltk
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.text import Tokenizer
from nltk.corpus import stopwords
import pickle
```

```python
df = pd.read_excel("data_final.xlsx")
```

```python
#data cleaning
def review_cleaning(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', ' ', text)
    text = re.sub('https?://\S+|www\.\S+', ' ', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text) 
    text = re.sub('\n', ' ', text)
    text = re.sub('\w*\d\w*', ' ', text)
    text = re.sub("<[a][^>]*>(.+?)</[a]>", 'Link.', text)
    text = re.sub('&gt;', " ", text)
    text = re.sub('&#x27;', "'", text)
    text = re.sub('&#x2F;', ' ', text)
    text = re.sub('<p>', ' ', text)
    text = re.sub('<i>', ' ', text)
    text = re.sub('</i>', ' ', text) 
    text = re.sub('&#62;', ' ', text)
    text = re.sub("\n", ' ', text)
    return text
```

```python
#apply data cleaning function
df['comment']=df['comment'].apply(lambda x:review_cleaning(x))
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
model = load_model("transformer_classification_sentiment_model.h5")

with open("tokenizer.pickle", "rb") as handle:
    tokenizer = pickle.load(handle)
```

```python
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

def predict_sentiment(text):
    cleaned = preprocess_text(text)
    sequence = tokenizer.texts_to_sequences([cleaned])
    padded = pad_sequences(sequence, maxlen=100, padding='post', truncating='post')
    prediction = model.predict(padded)[0][0]
    
    if 0.4 <= prediction <= 0.6:
        return "Neutral"
    elif prediction > 0.6:
        return "Positive"
    else:
        return "Negative"
```

```python
df['predicted_sentiment'] = df['comment'].apply(predict_sentiment)

print(df)
```
