```python
def extract_video_id(url):
    """Extract video ID from a YouTube URL or Shorts."""
    pattern = r"(?:v=|youtu\.be/|shorts/)([0-9A-Za-z_-]{11})"
    match = re.search(pattern, url)
    return match.group(1) if match else None

def clean_text(text):
    """Remove illegal characters for Excel export."""
    return re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f]', '', text)

def get_comments(video_id, category, part="snippet", max_results=100000):
    """Get comments from a YouTube video and tag them with category."""
    youtube = build("youtube", "v3", developerKey=DEVELOPER_KEY)
    comments = []
    total_fetched = 0

    try:
        request = youtube.commentThreads().list(
            part=part,
            videoId=video_id,
            textFormat="plainText",
            maxResults=100  # YouTube API max per page
        )

        while request and total_fetched < max_results:
            response = request.execute()
            for item in response.get("items", []):
                comment_text = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
                likes = item["snippet"]["topLevelComment"]["snippet"]["likeCount"]
                comments.append({
                    "video_id": video_id,
                    "comment": clean_text(comment_text),
                    "num_of_likes": likes,
                    "categorie": category
                })
                total_fetched += 1
                if total_fetched >= max_results:
                    break
            request = youtube.commentThreads().list_next(request, response)

        return comments

    except HttpError as error:
        print(f"HTTP error {error.status_code} for video {video_id}: {error}")
        return []

def main():
    all_comments = []

    all_sources = [
        ("food", list_url_food),
        ("sleeping", list_sleeping_massage),
        ("makeup", list_makeup)
    ]

    for category, url_list in all_sources:
        print(f"\nProcessing category: {category}")
        for url in url_list:
            video_id = extract_video_id(url)
            if not video_id:
                print(f"Invalid YouTube URL: {url}")
                continue

            print(f"  → Video: {video_id}")
            comments = get_comments(video_id, category)

            if comments:
                all_comments.extend(comments)
            else:
                print(f"  No comments found for video {video_id}")

    if all_comments:
        df = pd.DataFrame(all_comments)
        df = df.sort_values(by=["categorie", "num_of_likes"], ascending=[True, False])
        df.to_excel("Youtube_comments.xlsx", index=False)
        print("\n✅ All comments saved to Youtube_comments.xlsx")
    else:
        print("\n⚠️ No comments collected.")

if __name__ == "__main__":
    main()
```
